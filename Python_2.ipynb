```python
#Para poder usar dataframe vamos a activar dos bibliotecas. Una de ella es Numpy y la otra es pandas. Las dos se complementan y hacen cosas completamente diferentes.
#Para crear dataframe existen varias posibilidades, sin embargo, la más usada es la lectura directa desde un archivo.
#Opción 1:
import pandas as pd
import numpy as np
```

```python
#La primera forma de crear un dataframe es usando el método array, y luego guardarlo
dataframe_1= np.array([[2,4], [6,8], [10,12], [14,16]])
```

```python
#Para crear el dataframe del arreglo anterior debemos usar pandas
pd.DataFrame(dataframe_1)
```

```python
# Crear una tabla similar a esta pero de 5 en 5 hasta 50
dataframe_2= np.array([[5,10], [15,20], [25,30], [35,40], [45,50]])
pd.DataFrame(dataframe_2)
```

```python
#Crear una tabla similiar a la anteior pero de 3 en 3 hasta 30
dataframe_3= np.array([[3,6,9], [12,15,18], [21,24,27], [30,0,0]])
pd.DataFrame(dataframe_3)
```

```python
#Ahora vamos a cambiar el nombre a las columnas. Esto se hace de la siguiente manera:
df1=pd.DataFrame(dataframe_1, index=["fila1", "Fila2", "Fila3", "Fila4"],
                 columns=["Columna1", "Columna2"])
df1
```

```python
#Crear una tabla con información de 4 personas
dataframe_5=[["María", "Alexandra", "Escobar", 28], ["María", "Vannesa", "Murillo",22], ["Sebastián", "Melo", "Arroyo", 27], ["Miguel", "Angel", "Duran", 27]]
```

```python
#Para el tercer método uso los direccionarios de la siguiente manera
#Creamos 2 listas. Puedo crear cuantas listas quiera entendiendo que las listas son las columnas del DataFrame
CiudadColombia=["Bogotá", "Medellín", "Cali","Barranquilla"]
CiudadesInformación=[11000000,2500000,2200000,1300000]
```
