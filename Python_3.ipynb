```python
# Siempre, lo primero que debo hacer es importar Pandas y Numpy
import pandas as pd
import numpy as np
```

```python
#Ahora subimos el archivo desde google drive sin looguearnos
df_estudiantes=pd.read_csv("/content/drive/MyDrive/StudentsPerformance (1).csv")
df_estudiantes
```

```python
#Sigamos con la creaciÃ³n de columnas usando el mÃ©todo assign o el mÃ©todo insert
#Si quiero agregar varias columnas uso assign y tambiÃ©n si quiero sobreescribir una columna
filosofia_score=np.random.randint(0,100, size=1000)
fisica_score=np.random.randint(0,100, size=1000)
```

```python
#Ahora le pondrÃ© un indÃ­ce a cada fila para que sepa donde va en el DataFrame
filosofia_nota=pd.Series(filosofia_score, index=np.arange(0,1000))
fisica_nota=pd.Series(fisica_score, index=np.arange(0,1000))
```

```python
#Ahora lo que voy a hacer es poner las nuevas columnas a mi DataFrame, usando asing
df_estudiantes.assign(filosofia_score=filosofia_nota, fisica_score=fisica_nota)
df_estudiantes
```

```python
df_estudiantes=df_estudiantes.assign(filosofia_score=filosofia_nota, fisica_score=fisica_nota)
df_estudiantes
```

```python
#Para usar el mÃ©todo insert este no me devuelve copias sino que el cambio queda guardado de inmediato
df_estudiantes.insert(1,"Ed. Score", fisica_nota)
df_estudiantes
```
